# Creating a Conditional Column from 2 Choices: np.where()

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive')
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Inexpensive')
```

```python
# show first 5 rows
df_laptops.head()
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```
