```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

# isin(): Single Filtering

```python
# select Apple or HP Laptops
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframes
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])]
```

# isin(): Multiple Filtering

```python
# find notebooks or ultabooks from Apple or HP
filter1 = df_laptops['Company'].isin(['Apple', 'HP'])
filter2 = df_laptops['TypeName'].isin(['Notebook', 'Ultrabook'])
```

```python
# filter dataframe
df_laptops[filter1 & filter2]
```
